# Forming a Line Plot

```r
x = c(1,2,3,4,5)
y = c(10,7,9,8,6)

plot(x, y, type="l")
```

# Graph Colors

```r
x = c(1,2,3,4,5)
y = c(10,7,9,8,6)

plot(x, y, type="l", col="blue")                   #col= ["red", "green", "yellow","black", "blue", "white", "orange", "violet", "gray"
```

# Line width

```r
x = c(1,2,3,4,5)
y = c(10,7,9,8,6)

plot(x, y, type="l", col="blue", lwd=2) 
```

# Line Styles

Available parameter values for lty:

0 removes the line

1 displays a solid line

2 displays a dashed line

3 displays a dotted line

4 displays a "dot dashed" line

5 displays a "long dashed" line

6 displays a "two dashed" line

```r
x = c(1,2,3,4,5)
y = c(10,7,9,8,6)

plot(x, y, type="l", col="blue", lwd=5, lty=1, main = "lty = '1'") 
plot(x, y, type="l", col="blue", lwd=5, lty=2, main = "lty = '2'") 
plot(x, y, type="l", col="blue", lwd=5, lty=3, main = "lty = '3'") 
plot(x, y, type="l", col="blue", lwd=5, lty=4, main = "lty = '4'") 
plot(x, y, type="l", col="blue", lwd=5, lty=5, main = "lty = '5'") 
plot(x, y, type="l", col="blue", lwd=5, lty=6, main = "lty = '6'") 
```

# Multiple Lines

```r
line1 <- c(1,2,3,4,5,10)
line2 <- c(2,5,7,8,9,10)

plot(line1, type = "l", col = "blue")
lines(line2, type="l", col = "red")
```

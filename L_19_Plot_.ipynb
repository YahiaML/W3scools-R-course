# Forming a Plot

```r
x = c(1,2,3,4,5)
y = c(10,7,9,8,6)
plot(x,y)
```

# Plot Types

```r
x = c(1,2,3,4,5)
y = c(10,7,9,8,6)
plot(x, y, type = "l", main = "type = 'l'")
plot(x, y, type = "s", main = "type = 's'")
plot(x, y, type = "p", main = "type = 'p'")
plot(x, y, type = "o", main = "type = 'o'")
plot(x, y, type = "b", main = "type = 'b'")
plot(x, y, type = "h", main = "type = 'h'")
```

# Plot Labels

```r
x = c(1,2,3,4,5)
y = c(10,7,9,8,6)
plot(x, y, main="My Graph", xlab="The x-axis", ylab="The y axis")
```

# Size

```r
x = c(1,2,3,4,5)
y = c(10,7,9,8,6)
plot(x, y, cex=2, main="My Graph", xlab="The x-axis", ylab="The y axis",
     cex.main = 2,   # Title size
     cex.sub = 1.5,  # Subtitle size
     cex.lab = 3,    # X-axis and Y-axis labels size
     cex.axis =1.5) # Axis labels size
```

# Point Shapes

r_plot_pch2.png

```r
x = c(1,2,3,4,5)
y = c(10,7,9,8,6)
plot(x, y, pch=19, cex=2)
```

# Graph Colors

```r
x = c(1,2,3,4,5)
y = c(10,7,9,8,6)


par(bg = 'gray')            # background color

plot(x, y, pch=19, cex=2, col="red",  #col= ["red", "green", "yellow","black", "blue", "white", "orange", "violet", "gray"
    main="My Graph", xlab="The x-axis", ylab="The y axis",
    col.main = "green",     # Title color
    col.sub = "blue",       # Subtitle color
    col.lab = "sienna2",    # X and Y-axis labels color
    col.axis = "maroon4",   # Tick labels color
    fg = "orange")          # Box color
```

# Box shape

```r
x = c(1,2,3,4,5)
y = c(10,7,9,8,6)

plot(x, y, bty = "o", main = "Default")
plot(x, y, bty = "7", main = "bty = '7'")
plot(x, y, bty = "L", main = "bty = 'L'")
plot(x, y, bty = "U", main = "bty = 'U'")
plot(x, y, bty = "C", main = "bty = 'C'")
plot(x, y, bty = "n", main = "bty = 'n'")

par(mfrow = c(1, 1))
```

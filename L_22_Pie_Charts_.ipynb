# Forming a Pie Chart

```r
# Create a vector of pies

x <- c(10,50,5,90)            # The size of each pie is determined by comparing the value with all the other values, by using this formula: x/sum(x)

# Display the pie chart
pie(x)
```

# Start Angle

```r
# Create a vector of pies
x <- c(10,50,5,90)

# Display the pie chart and start the first pie at 90 degrees
pie(x, init.angle = 90)
```

# Labels and Header

```r
# Create a vector of pies
x <- c(5, 50, 10, 90)

# Create a vector of labels
mylabel <- c("Dates", "Bananas", "Cherries", "Apples")

# Display the pie chart with labels
pie(x, label = mylabel, main = "Fruits")
```

# Colors

```r
# Create a vector of pies
x <- c(5, 50, 10, 90)

# Create a vector of labels
mylabel <- c("Dates", "Bananas", "Cherries", "Apples")

colors <- c("blue", "yellow", "green", "orange")

# Display the pie chart with labels
pie(x, label = mylabel, main = "Fruits", col = colors)
```

# Legend

```r
# Create a vector of pies
x <- c(5, 50, 10, 90)

# Create a vector of labels
mylabel <- c("Dates", "Bananas", "Cherries", "Apples")

colors <- c("blue", "yellow", "green", "orange")

# Display the pie chart with labels
pie(x, label = mylabel, main = "Fruits", col = colors)
# Display the explanation box
legend("bottomright", mylabel, fill = colors)  # bottomright, bottom, bottomleft, left, topleft, top, topright, right, center
```

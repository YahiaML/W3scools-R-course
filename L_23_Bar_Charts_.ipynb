# Forming a Bar Chart

```r
# x-axis values
x <- c("Mostafa", "Yahia", "Abd Allah", "Ali")

# y-axis values
y <- c(25, 21, 18, 15)

barplot(y, names.arg = x)     # names.arg defines the name
```

# Colors

```r
# x-axis values
x <- c("Mostafa", "Yahia", "Abd Allah", "Ali")

# y-axis values
y <- c(25, 21, 18, 15)

barplot(y, names.arg = x, col = "Orange")    
```

# Density / Bar Texture

```r
# x-axis values
x <- c("Mostafa", "Yahia", "Abd Allah", "Ali")

# y-axis values
y <- c(25, 21, 18, 15)

barplot(y, names.arg = x, density = 5)  
barplot(y, names.arg = x, density = 10)    
barplot(y, names.arg = x, density = 15)    
```

# Bar Width

```r
x <- c("Mostafa", "Yahia", "Abd Allah", "Ali")
y <- c(25, 21, 18, 15)

barplot(y, names.arg = x, width = c(4,3,2,1))   
```

# Horizontal Bars

```r
# x-axis values
x <- c("Mostafa", "Yahia", "Abd Allah", "Ali")

# y-axis values
y <- c(25, 21, 18, 15)

barplot(y, names.arg = x, horiz = TRUE)     
```
